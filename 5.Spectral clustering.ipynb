```python
#importing the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
#using iloc column position name we can access, columnwise select using for iloc.
X = dataset.iloc[:,[3,4]].values
#dataset.iloc [:,-(row path), it's refereing all rows in this set taking and [3,4] means column wise 3 and 4 th column values.
#Using above line we can split or take it seperately on columns
```

```python
X
```

```python
from sklearn.cluster import SpectralClustering
clusmodel = SpectralClustering(n_clusters=3,assign_labels='discretize', random_state=0)
label = clusmodel.fit_predict(X)
```

```python
clusmodel
```

```python
label
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['Cluster_group']=label
```

```python
supervised
```

```python
import seaborn as sns #seaborn  library using for graph purpose
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], #sns.lmplot using lmplot we can do it easily,here we are passing datala supervised data, X-axis we need x=supervised.columns[3] and y-axis we passing y=supervised.columns[4], and hue=supervised.columns[5] using to diplay colours for all clusters here 5 cluster means it'll show all cluster group in 5 colours(5 coulmn it'll show in 5 different colours) 
                   fit_reg=False, legend=True, legend_out=True) #fit_reg=False here we are not doing any regression thats why mentioned like false.#legend=True means it'llshow like 0 means blue color category/cluster and 1 means orange color category like that it'll show
```
